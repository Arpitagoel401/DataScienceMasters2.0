#### Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps with the highest number of installs.

```python
df['Installs'] = pd.to_numeric(df['Installs'].str.replace(',', '').str.replace('+', ''), errors='coerce')

df = df.dropna(subset=['Installs'])

app_installs_frequency = df.groupby('App')['Installs'].sum().reset_index()

app_installs_frequency = app_installs_frequency.sort_values(by='Installs', ascending=False)

# Display the top 10 most popular apps
top_10_apps = app_installs_frequency.head(10)
print(top_10_apps)
```

#### Q17. A marketing research firm wants to analyze the Google Playstore dataset to identify the best time to launch a new app. What features would you analyze to make recommendations to the company, and what data visualizations would you use to present your findings?

```python
time_analysis = df.copy()
time_analysis['Last Updated'] = pd.to_datetime(time_analysis['Last Updated'])
time_analysis.set_index('Last Updated', inplace=True)
time_analysis.resample('M')['Rating'].mean().plot()
plt.title('Average Ratings Over Time')
```
