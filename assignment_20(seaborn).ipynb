Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

```python
# Scatter Plot:Visualize the relationship between two numerical variables.
# Example: Displaying the correlation between a student's study hours and exam scores.
    
# Histogram:Illustrate the distribution of a single numerical variable.
# Example: Showing the frequency distribution of ages in a population.
    
# Bar Plot:Compare the values of different categories in a dataset.
# Example: Comparing sales performance across different product categories.
    
# Box Plot (Box-and-Whisker Plot):Summarize the distribution of a numerical variable, highlighting outliers.
# Example: Analyzing the spread of income levels in different regions.
    
# Heatmap:Display the correlation matrix or intensity of relationships in a dataset.
# Example: Visualizing the correlation between various factors in a scientific experiment.
```

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
fmri = sns.load_dataset("fmri")
fmri.head()
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="region", style="event")
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the titanic dataset
titanic_data = sns.load_dataset("titanic")

# Set up the matplotlib figure
plt.figure(figsize=(12, 6))

# Plot the first box plot for 'age'
plt.subplot(1, 2, 1)
sns.boxplot(x='pclass', y='age', data=titanic_data)
plt.title('Box Plot of Age by Pclass')

# Plot the second box plot for 'fare'
plt.subplot(1, 2, 2)
sns.boxplot(x='pclass', y='fare', data=titanic_data)
plt.title('Box Plot of Fare by Pclass')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights_data = sns.load_dataset("flights")

# Pivot the dataset to create a matrix suitable for a heatmap
flights_pivot = flights_data.pivot_table(index='month', columns='year', values='passengers')

# # Set up the matplotlib figure
plt.figure(figsize=(12, 8))

# Plot the heatmap
sns.heatmap(flights_pivot, cmap='viridis', annot=True, fmt='d', linewidths=.5)

# Set plot labels and title
plt.xlabel("Year")
plt.ylabel("Month")
plt.title("Passenger Counts by Month and Year (Flights Dataset)")

# Show the plot
plt.show()
```
