Q6. What do you understand about the windows function in pandas and list the types of windows
functions?

```python
# In pandas, window functions are operations that operate on a set of data points defined by a window of fixed size or rolling 
# nature. These functions are especially useful for time-series data and other ordered data sets. 
df4 = pd.DataFrame({'a':[1,2,3,4,5,6,7,8,9]})
# Rolling Windows:
# rolling(): This function is used to provide rolling view of the data. It creates an object that enables rolling window 
# calculations. You can then apply various aggregation functions (like mean, sum, etc.)
print(df4.rolling(window=2).mean())
# expanding(): This function provides expanding window calculations.
print(df4.expanding().sum())
#cumsum()
print(df4.cumsum())
```

Q9. Write a Python program that reads a CSV file containing categorical data and converts a specified
column to a categorical data type. The program should prompt the user to enter the file path, column
name, and category order, and then display the sorted data.

```python
df = pd.read_csv("services.csv")
print(df.head())
df['status'] = pd.Categorical(df['status'], categories=None, ordered=True)
sorted_df = df.sort_values(by=['status'])
print("\nSorted Data:")
print(sorted_df)
```

Q11. You are given a CSV file containing student data that includes the student ID and their test score. Write
a Python program that reads the CSV file, calculates the mean, median, and mode of the test scores, and
displays the results in a table.
The program should do the following
I Prompt the user to enter the file path of the CSV file containing the student data
I Read the CSV file into a Pandas DataFrame
I Calculate the mean, median, and mode of the test scores using Pandas tools
I Display the mean, median, and mode in a table.

```python
df = pd.read_csv('student_data.csv')
mean_score = df['Test Score'].mean()
median_score = df['Test Score'].median()
mode_score = df['Test Score'].mode().values[0]

# Display the results in a table
result_table = pd.DataFrame({
    'Statistic': ['Mean', 'Median', 'Mode'],
    'Test Score': [mean_score, median_score, mode_score]
})

print(result_table)
```
